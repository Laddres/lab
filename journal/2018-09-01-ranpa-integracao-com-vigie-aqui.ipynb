```python
def padronizarCPF(valor):
    if valor is None:
        return None
    
    numeros = ''.join(re.findall(r'\d+', valor))
    
    if len(numeros) == 11:
        return numeros
    
    if len(numeros) > 11:
        return numeros[:11]

    return padronizarCPF('0' + numeros)
```

```python
print(padronizarCPF('517934821-87'))

print(padronizarCPF('7039625903'))

print(padronizarCPF('7593929'))

print(padronizarCPF(None))

print(padronizarCPF('564.611.350-490'))
```

```python
def padronizarDescricao(valor):
    if valor is None:
        return None

    return re.sub(r'([\n\r]|&nbsp;)', '', valor.strip())
```

```python
df = pd.DataFrame()
```

```python
for candidato in json.itertuples():
    numero_processos = len(candidato.data['lawsuits'])
    if (numero_processos == 0):
        df_i = pd.DataFrame(data = {
            'id_politico': [candidato.data['id']],
            'nome': [candidato.data['names'][0]],
            'nome_urna': [candidato.data['display_name']],
            'cpf': [padronizarCPF(candidato.data['cpf'])],
            'processo_id': None,
            'processo_tipo': None,
            'processo_numero': None,
            'processo_tribunal': None,
            'processo_descricao': None,
            'processo_link': None,
            'processado': [candidato.data['purple']],
        })
        df = df.append(df_i)
        continue
    
    for processo in candidato.data['lawsuits']:
        df_i = pd.DataFrame(data = {
            'id_politico': [candidato.data['id']],
            'nome': [candidato.data['names'][0]],
            'nome_urna': [candidato.data['display_name']],
            'cpf': [padronizarCPF(candidato.data['cpf'])],
            'processo_id': [processo['id']],
            'processo_tipo': [processo['type']],
            'processo_numero': [processo['number']],
            'processo_tribunal': [processo['department']],
            'processo_descricao': [padronizarDescricao(processo['description']['pt-br'])],
            'processo_link': [processo['link']],
            'processado': [candidato.data['purple']],
        })
        df = df.append(df_i)
```
