## Atualizar Fotos no Banco de Dados

```python
import sys
import mysql.connector as mysql

BD_CONFIG = {
    'user': 'root',
    'password': '',
    'host': '127.0.0.1',
    'database': 'laddres'
}
```

```python
cnx = mysql.connect(**BD_CONFIG)
cursor = cnx.cursor()

total = len(df_fotos)
for candidato in df_fotos.itertuples():
    sys.stdout.write('\r{} de {}'.format(candidato.Index, total))
    sys.stdout.flush()

#     query = 'UPDATE candidato SET foto = "{}" WHERE id = {}'.format(candidato.foto, candidato.candidato_id)
#     query = 'UPDATE candidato SET foto = %s WHERE id = %s'
    query = (
        "UPDATE candidato "
        "SET "
        "foto = %(foto)s "
        "WHERE id = %(candidato_id)s"
    )
    atributos = {
        'foto': candidato.foto,
        'candidato_id': candidato.candidato_id
    }

    try:
#         cursor.execute(query, (candidato.foto, candidato.candidato_id))
        cursor.execute(query, atributos)
        cnx.commit()
    except:
        print(candidato)
        cursor.close()
        cnx.close()
        raise

cursor.close()
cnx.close()
```
