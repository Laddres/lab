```python
import pandas as pd
import requests
from utils.camara_obter_dados import obter_dados
from utils.camara_obter_dados import obter_votos
```

```python
# Pretty Print
def pp(df):
    for row in df.itertuples():
        print('{} {}/{} - {} ({})\n'.format(row.siglaTipo, row.numero, row.ano, row.ementa, row.id))
```

```python
# Check se há votações para estas proposições
def check(df):
    for row in df.itertuples():
        url = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes/{}/votacoes'.format(row.id)
        request = requests.get(url)
        if request.status_code == 200:
            response = request.json()
            if (len(response['dados'])):
                print('{} - {} votações'.format(row.id, len(response['dados'])))
        else:
            print('Erro ao tentar carregar dados da API (id: {})'.format(row.id))
```
