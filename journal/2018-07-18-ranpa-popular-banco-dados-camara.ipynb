```python
import pandas as pd
import numpy as np
import sys
import mysql.connector as mysql

BD_CONFIG = {
    'user': 'root',
    'password': '',
    'host': '127.0.0.1',
    'database': 'laddres'
}

def migrar_dados(dataframe, tabela, colunas_tabela, colunas_dataframe):
    cnx = mysql.connect(**BD_CONFIG)
    cursor = cnx.cursor()

    total = len(dataframe)
    for row in dataframe.itertuples():
        columns = ', '.join(colunas_tabela)
        placeholders = ', '.join(np.full((1, len(colunas_tabela)), '%s')[0])

        sql = 'INSERT INTO {} ({}) VALUES ({})'.format(tabela, columns, placeholders)
        values = []
        for column in colunas_dataframe:
            values.append(getattr(row, column))

        try:
            cursor.execute(sql, tuple(values))
            cnx.commit()
        except:
            print(row)
            cursor.close()
            cnx.close()
            raise

        sys.stdout.write('\r{} de {}'.format((row.Index + 1), total))
        sys.stdout.flush()


    cursor.close()
    cnx.close()
```

```python
migrar_dados(
    dataframe=legislaturas,
    tabela='camara_legislatura',
    colunas_tabela=['id', 'data_inicio', 'data_fim'],
    colunas_dataframe=['id', 'dataInicio', 'dataFim']
)
```

Todos os deputados desse dataset tem pelo menos:

- Nome
- Nome Civil
- Data de nascimento
- Partido
- Sgila UF

A ideia é buscar no banco de dados pelo `nomeCivil` e `dataNascimento`

```python
def _get_id_candidato(nome, data_nascimento):
    cnx = mysql.connect(**BD_CONFIG)
    cursor = cnx.cursor()

    query = ("SELECT id FROM candidato WHERE "
             "nome = %(nomeCivil)s AND data_nascimento = %(dataNascimento)s")
    parametros = {
        'nomeCivil': nome,
        'dataNascimento': data_nascimento
    }

    try:
        cursor.execute(query, parametros)
    except:
        print(query)
        cursor.close()
        cnx.close()
        raise

    resultados = cursor.fetchall()
    
    if (len(resultados) > 1):
        print('Mais de um deputado com o mesmo nome: {}'.format(nome))

    id = resultados[0][0] if len(resultados) > 0 else  None

    cursor.close()
    cnx.close()

    return id
```

```python
# Gerar o dataframe que será usado pra popular a tabela 'deputado' do banco de dados
def gerar_dataframe_deputado(dataframe):
    total = len(dataframe)
    deputados_df = pd.DataFrame()
    for deputado in dataframe.itertuples():
        idCandidato = _get_id_candidato(deputado.nomeCivil, deputado.dataNascimento)

        df_i = pd.DataFrame([[
            deputado.id,
            idCandidato,
            deputado.dataUltimoStatus,
            deputado.gabinete_nome,
            deputado.gabinete_predio,
            deputado.gabinete_sala,
            deputado.gabinete_andar,
            deputado.gabinete_telefone,
            deputado.gabinete_email,
            deputado.urlWebsite
        ]],
        columns=[
            'id',
            'candidato_id',
            'data_ultimo_status',
            'gabinete_nome',
            'gabinete_predio',
            'gabinete_sala',
            'gabinete_andar',
            'gabinete_telefone',
            'gabinete_email',
            'website'
        ])
        deputados_df = deputados_df.append(df_i)

        sys.stdout.write('\r{} de {}'.format((deputado.Index + 1), total))
        sys.stdout.flush()

    return deputados_df
```
