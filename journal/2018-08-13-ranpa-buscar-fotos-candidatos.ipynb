# SQL views e hot tables

## Dados Iniciais do Candidato

```SQL
CREATE OR REPLACE VIEW
  view_dados_candidato
AS
  SELECT
    candidatura.candidato_id as id,
    candidatoInno.nome as nome,
    candidatura.nome_urna as nomeUrna,
    candidatoInno.data_nascimento as dataNascimento,
    cidade.nome as cidadeNatal,
    estado.nome as estadoNatal,
    partido.sigla as partido,
    candidatura.numero_candidato as numero,
    candidatura.estado_id as idEstadoCandidatura,
    candidatura.cargo_id as idCargo,
    cargo.nome as cargo,
    grau_instrucao.descricao as grauInstrucao,
    ocupacao.descricao as ocupacao
  FROM
    candidatura
      LEFT JOIN partido on candidatura.partido_id = partido.id
      LEFT JOIN cargo on candidatura.cargo_id = cargo.id
      LEFT JOIN candidatoInno on candidatura.candidato_id = candidatoInno.id
      LEFT JOIN cidade on candidatoInno.cidade_id = cidade.id
      LEFT JOIN estado on cidade.estado_id = estado.id
      LEFT JOIN grau_instrucao on candidatoInno.grau_instrucao_id = grau_instrucao.id
      LEFT JOIN ocupacao on candidatoInno.ocupacao_id = ocupacao.id
  WHERE
    candidatura.eleicao_id = 138 AND
    candidatura.situacao_candidatura_id = 2

CREATE TABLE hot_dados_candidato AS SELECT * FROM view_dados_candidato;

CREATE INDEX hot_dados_candidato_id_index ON hot_dados_candidato (id);
CREATE INDEX hot_dados_candidato_nome_index ON hot_dados_candidato (nome);
CREATE INDEX hot_dados_candidato_nomeUrna_index ON hot_dados_candidato (nomeUrna);

DROP VIEW view_dados_candidato;
```

## Candidaturas Anteriores do Candidato (to improve)

```SQL
CREATE VIEW
  candidaturas_candidato
AS
  SELECT
    candidatura.id,
    candidatura.candidato_id,
    cargo.nome,
    partido.sigla,
    eleicao.ano as eleicao_ano,
    eleicao.descricao as eleicao_descricao,
    cidade.nome as cidade,
    estado.sigla as estado_sigla,
    estado.nome as estado_nome,
    candidatura.legenda_nome,
    candidatura.legenda_composicao,
    resultado_candidatura.descricao as resultado
  FROM
    candidatura
      LEFT JOIN cargo on candidatura.cargo_id = cargo.id
      LEFT JOIN partido on candidatura.partido_id = partido.id
      LEFT JOIN eleicao on candidatura.eleicao_id = eleicao.id
      LEFT JOIN cidade on candidatura.cidade_id = cidade.id
      LEFT JOIN estado on candidatura.estado_id = estado.id
      LEFT JOIN resultado_candidatura on candidatura.resultado_candidatura_id = resultado_candidatura.id
  WHERE
    candidatura.resultado_candidatura_id != 6
  ORDER BY
    candidatura.candidato_id ASC,
    candidatura.eleicao_id DESC
```
