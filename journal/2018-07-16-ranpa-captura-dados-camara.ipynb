## Deputados

Vamos buscar os deputados que participaram da câmara de deputados em cada uma das legislaturas que contém dados válidos (52 a 55)

```python
legislatura_atual = df_legislaturas.at[0, 'id']
legislaturas_validas = [str(legislatura) for legislatura in (range(52, legislatura_atual + 1, 1))]

legislaturas_validas
```

```python
import pandas as pd

df_deputados = pd.DataFrame()
for legislatura in legislaturas_validas:
    print('Legislatura: {}'.format(legislatura))
    df_deputados_i = obter_dados(
        endpoint='/deputados',
        filtros=[
            'itens=100',
            'idLegislatura={}'.format(legislatura)
        ]
    )
    df_deputados = df_deputados.append(df_deputados_i)
```

```python
df_deputados
```

```python
# this index contains no useful information
df_deputados.index = df_deputados.reset_index().index
```

```python
dataframe_to_file(
    dataframe=df_deputados,
    nome='deputados',
    origem_dados='camara'
)
```

```python
df_deputados.sort_values(by=['nome'], ascending=True, na_position='first')
```

```python
import pandas as pd
from utils.camara_obter_dados import obter_dados
from utils.dataframe_to_file import dataframe_to_file

orgaos = pd.read_csv('../data/2018-07-16-camara-orgaos')

membros_orgaos_df = pd.DataFrame()
for orgao in orgaos.itertuples(index=True):
    print('\nÓrgão {}: {} de {}'.format(orgao.id, orgao[0], len(orgaos)))

    df_i = obter_dados(
        endpoint='/orgaos/{}/membros'.format(orgao.id),
        filtros=['itens=100']
    )
    
    if len(df_i) == 0:
        continue
    
    df_i['idOrgao'] = orgao.id
    membros_orgaos_df = membros_orgaos_df.append(df_i[[
        'id',
        'idLegislatura',
        'nomePapel',
        'idPapel',
        'dataInicio',
        'dataFim'
    ]])

dataframe_to_file(
    dataframe=membros_orgaos_df,
    nome='membros-orgaos',
    origem_dados='camara'
)
```
