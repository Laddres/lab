```python
def padronizarCPF(valor):
    if valor is None:
        return None
    
    numeros = ''.join(re.findall(r'\d+', valor))
    
    if len(numeros) == 11:
        return numeros
    
    if len(numeros) > 11:
        return numeros[:11]

    return padronizarCPF('0' + numeros)
```

```python
def padronizarTituloEleitoral(valor):
    if valor is None:
        return None
    
    numeros = ''.join(re.findall(r'\d+', valor))

    if len(numeros) == 12:
        return numeros

    return None
```

```python
def padronizarDescricao(valor):
    if valor is None:
        return None

    return re.sub(r'([\n\r]|&nbsp;)', '', valor.strip())
```

```python
def _get_candidato_id(politico, cnx, cursor):
    if (politico.cpf is not None):
        # query = 'SELECT id FROM candidato WHERE cpf = "{}"'.format(politico.cpf)
        query = 'SELECT id FROM candidato WHERE cpf = "{}" ORDER BY id DESC LIMIT 1'.format(politico.cpf)
        cursor.execute(query)
        resultados = cursor.fetchall()
        
        if (len(resultados) == 1):
            return resultados[0][0]
        else:
            query = 'SELECT id FROM candidato WHERE nome = "{}"'.format(politico.nome)
            cursor.execute(query)
            resultados = cursor.fetchall()

            if (len(resultados) == 1):
                return resultados[0][0]
    
    if (politico.titulo_eleitoral is not None):
        query = 'SELECT id FROM candidato WHERE titulo_eleitoral = "{}"'.format(politico.titulo_eleitoral)
        cursor.execute(query)
        resultados = cursor.fetchall()
        
        if (len(resultados) == 1):
            return resultados[0][0]
    
    query = 'SELECT id FROM candidato WHERE nome = "{}"'.format(politico.nome)
    cursor.execute(query)
    resultados = cursor.fetchall()
    
    if (len(resultados) == 1):
        return resultados[0][0]
    
    if (politico.contato is not None and re.search('(@camara.leg.br)', politico.contato)):
        query = 'SELECT candidato_id FROM deputado WHERE gabinete_email = "{}"'.format(politico.contato.strip())
        cursor.execute(query)
        resultados = cursor.fetchall()

        if (len(resultados) == 1):
            return resultados[0][0]
    
    return None
```
