```python
import re

def get_page_number(url):
    pattern = r'pagina=(\d+)'
    match = re.findall(pattern, url)

    return 1 if len(match) == 0 else match[0]

def get_last_page_number(links):
    for link in links:
        if link['rel'] == 'last':
            return get_page_number(link['href'])
```

```python
import pandas as pd
import requests

url_base = 'https://dadosabertos.camara.leg.br/api/v2'
url_legislaturas = '{}/legislaturas?pagina='.format(url_base)

df = pd.DataFrame()
pagina = 1
while (pagina > 0):
    r = requests.get('{}{}'.format(url_legislaturas, pagina))
    if r.status_code == 200:
        json = r.json()
        dados = json['dados']
        links = json['links']
        ultima_pagina = int(get_last_page_number(links))

        df = df.append(pd.DataFrame(dados), ignore_index = True)

        print('Página {} de {}'.format(pagina, ultima_pagina))
        print('Página {} processada com sucesso'.format(str(pagina)))

        if (pagina == ultima_pagina):
            pagina = -1
        else:
            pagina = pagina + 1
```
